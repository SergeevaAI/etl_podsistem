Этот шаг позволяет вычислить значения для определенной группы полей. 

image.png

Свойства:
1. Step name. Название этапа; это имя должно быть уникальным для одного преобразования. 
2. Include all rows? Включите, если вы хотите, чтобы в выходных данных отображались все строки, а не только агрегация.
3. Temporary files directory. Каталог, в котором хранятся временные файлы.
4. TMP-file prefix. 
5. Add line number, restart in each group. Позволяет добавлять номер строки, который перезапускается с 1 в каждой группе.
6. Line number field name. Позволяет добавлять номер строки, который перезапускается с 1 в каждой группе.
7. Always give back a row. Если вы включите эту опцию, пошаговая группировка всегда будет возвращать строку результатов, даже если строки ввода нет.
8. Group fields table. Укажите поля, по которым вы хотите сгруппировать.
